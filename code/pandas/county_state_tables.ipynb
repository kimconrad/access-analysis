## JOIN TABLES TO CREATE COUNTY_DATA_DF

```python
# Add total population column
# Join county_state_df and total_pop_df
county_data_df = county_state_df.join(total_pop_df.set_index("id"), on="id", how="left")
county_data_df
```

```python
# drop extra column
county_data_df.drop(["Geographic Area Name"], axis=1, inplace=True)
county_data_df
```

```python
# Add total households and avg income columns
# Join county_data_df and avg_income_df
county_data_df = county_data_df.join(avg_income_df.set_index("id"), on="id", how="left")
county_data_df
```

```python
# Drop geographic_area_name column
county_data_df.drop(["geographic_area_name"], axis=1, inplace=True)
county_data_df
```

```python
# add rural/urban column
# Join county_data_df and oregon_df
county_data_df = county_data_df.join(oregon_df.set_index("id"), on="id", how="left", rsuffix="_o")
county_data_df
```

```python
# Drop unnecessary columns
county_data_df.drop(["avg_household_income_dollars_o", "county_o", "state_o", "desktop_laptop", "perc_desktop_laptop", "broadband", "perc_broadband", "total_households"], axis=1, inplace=True)
county_data_df
```

```python
# add computer types
# join county_data_df and comp_types
county_data_df = county_data_df.join(comp_types_df.set_index("id"), on="id", how="left", rsuffix="_ct")
county_data_df
```

```python
# drop unnecessary columns
county_data_df.drop(["total", "geographic_area_name"], axis=1, inplace=True)
county_data_df
```

```python
# add income levels
# create a new df
income_df = pd.DataFrame()

# append rows

income_df["id"] = inc_int_comp_df["id"]
income_df["less10k"] = inc_int_comp_df["less_10000_total"]
income_df["10k"] = inc_int_comp_df["_10000_19900_total"]
income_df["20k"] = inc_int_comp_df["_20000_34999_total"]
income_df["35k"] = inc_int_comp_df["_35000_49999_total"]
income_df["50k"] = inc_int_comp_df["_50000_74900_total"]
income_df["greater75k"] = inc_int_comp_df["greater_75000_total"]
income_df
```

```python
# join county_data_df and inc_int_comp_df
county_data_df = county_data_df.join(income_df.set_index("id"), on="id", how="left")
county_data_df
```

```python
# add internet types
# create a new df
internet_df = pd.DataFrame()

# append rows

internet_df["id"] = inc_int_comp_df["id"]
internet_df["dialup"] = inc_int_comp_df["less_10000_dialup"] + inc_int_comp_df["_10000_19900_dialup"] + inc_int_comp_df["_20000_34999_dialup"] + inc_int_comp_df["_35000_49999_dialup"] + inc_int_comp_df["_50000_74900_dialup"] + inc_int_comp_df["greater_75000_dialup"] 
internet_df["broadband"] =  inc_int_comp_df["less_10000_broadband"] + inc_int_comp_df["_10000_19900_broadband"] + inc_int_comp_df["_20000_34999_broadband"] + inc_int_comp_df["_35000_49999_broadband"] + inc_int_comp_df["_50000_74900_broadband"] + inc_int_comp_df["greater_75000_broadband"] 
internet_df["internet"] = internet_df["dialup"] + internet_df["broadband"]
internet_df["no_internet"] = inc_int_comp_df["less_10000_none"] + inc_int_comp_df["_10000_19900_none"] + inc_int_comp_df["_20000_34999_none"] + inc_int_comp_df["_35000_49999_none"] + inc_int_comp_df["_50000_74900_none"] + inc_int_comp_df["greater_75000_none"] 
internet_df
```

```python
# join county_data_df and inc_int_comp_df
county_data_df = county_data_df.join(internet_df.set_index("id"), on="id", how="left")
county_data_df
```

```python
# add internet percentages
# create a new df
internet_perc_df = pd.DataFrame()

# append rows

internet_perc_df["id"] = internet_df["id"]
internet_perc_df["perc_dialup"] = internet_df["dialup"] / county_data_df["households"]
internet_perc_df["perc_broadband"] =  internet_df["broadband"] / county_data_df["households"]
internet_perc_df["perc_internet"] = internet_df["internet"] / county_data_df["households"]
internet_perc_df["perc_no_internet"] = internet_df["no_internet"] / county_data_df["households"]
internet_perc_df
```

```python
# join county_data_df and internet_perc_df
county_data_df = county_data_df.join(internet_perc_df.set_index("id"), on="id", how="left")
county_data_df
```

## FINISHING COUNTY_DATA_DF

```python
county_data_df.head()
```

```python
state_data_df["state"] = state_data_df.index
state_data_df.index = np.arange(0, len(state_data_df))
state_data_df.head()
```

```python
# Assign highest state income to each county

county_data_df["high_state_income"] = 0
county_data_df["low_state_income"] = 0
county_data_df
```

```python
# Adding highest state income to each county line

j = 0
for state in state_data_df["state"]:
    for i in range(len(county_data_df)):
        if county_data_df.iloc[i, 2] == state:
            county_data_df.iloc[i,31] = state_data_df.iloc[j,0]
    j = j+1
```

```python
# Adding lowest state income to each county line

j = 0
for state in state_data_df["state"]:
    for i in range(len(county_data_df)):
        if county_data_df.iloc[i, 2] == state:
            county_data_df.iloc[i,32] = state_data_df.iloc[j,1]
    j = j+1
```

```python
county_data_df
```

```python
# Create State Ratio Incomes
county_data_df["state_ratio_income"] = county_data_df["avg_household_income_dollars"] / county_data_df["high_state_income"]
county_data_df["state_ratio_income"] 
```

```python
# Create National Ratio Incomes
county_data_df["national_ratio_income"] = county_data_df["avg_household_income_dollars"] / county_data_df["high_state_income"].max()
county_data_df["national_ratio_income"] 
```
