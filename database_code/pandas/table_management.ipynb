```python
# Clean mean_income_df
avg_income_df = mean_income_df[["GEO_ID", "NAME", "S1902_C01_001E", "S1902_C03_001E"]]
avg_income_df = avg_income_df.rename(columns={"GEO_ID": "id", "NAME": "geographic_area_name", "S1902_C01_001E": "households", "S1902_C03_001E": "avg_household_income_dollars"})

avg_income_df
```

```python
# Check for null values
avg_income_df.isnull().values.any()
```

```python
# Replace id with numeric key
avg_income_df["id"].replace("\d{7}\D{2}", "", regex=True, inplace=True)
avg_income_df
```

```python
# Change datatype of id column
avg_income_df = avg_income_df.astype({"id": int})
avg_income_df.dtypes
```
